```python
#Importing the dataset and required libraries

import pandas as pd
df = pd.read_csv("Automobile_data.csv")
```

```python
#Printing the first 10 rows

df.head(30)
```

```python
#Printing the last 30 rows

df.tail(30)
```

```python
#Get some infos from the data

df.describe()
```

```python
#Get some infos for specific columns

df[['price','horsepower']].describe()
```

```python
#Clean data and update the CSV file

df["price"] = df["price"].fillna(df["price"].mean())

df.head(30)
```

```python
#Find the most expensive car company name

df1 = df[['company','price']][df.price==df['price'].max()]
df1
```

```python
#Print some German Cars details

ger_cars = df.loc[df['company'].isin(['volkswagen','audi','porsche','bmw','mercedes-benz'])]
ger_cars
```

```python
#Count total cars per company

df['company'].value_counts()
```

```python
#Count total cars per body-style

df['body-style'].value_counts()
```

```python
#Find each companys Higesht price car
car_Manufacturers = df.groupby('company')
priceDf = car_Manufacturers['company','price'].max()
priceDf
```

```python
#Find the average mileage of each car making company
mileageDf = car_Manufacturers['company','average-mileage'].mean()
mileageDf
```

```python
#Sorting Cars by price
carsDf = df.sort_values(by=['price'], ascending=False)
carsDf
```

```python
#Remove rows with non-priced cars

df1 = df[df.price != 0]
df1
```

```python
#Drawing a bar chart

import matplotlib.pyplot as plt


ax = df['company'].value_counts().plot(kind='bar', title="Number for each Car Company", 
                                       figsize=(18,8))
ax.set_xlabel("Car Company")
ax.set_ylabel("Total Number")
plt.show()
```

```python
#Drawing a bar chart

import matplotlib.pyplot as plt

color_list = ['firebrick', 'red', 'orange', 'green','magenta', 'khaki', 'blue','yellowgreen',
    'slateblue','black','tan','skyblue','lime','cyan','plum','gray']

ax = df['company'].value_counts().plot(kind='bar' , figsize=(18,8),fontsize=20, color = color_list)
ax.set_xlabel("Car Company",fontsize=20,color='blue')
ax.set_ylabel("Total Number",fontsize=20,color='red')
plt.suptitle('Number for each Car Company', fontsize=30,color="green")
plt.show()
```

```python
import matplotlib.pyplot as plt

color_list = ['firebrick', 'red', 'orange', 'green','magenta']

ax = df['body-style'].value_counts().plot(kind='bar' , figsize=(18,10),fontsize=20, color = color_list)
ax.set_xlabel("Car Body Style",fontsize=20,color='blue')
ax.set_ylabel("Total Number",fontsize=20,color='red')
plt.suptitle('Count of available Car Body Style', fontsize=30)
plt.show()
```

```python
import matplotlib.pyplot as plt

color_list = ['firebrick', 'red', 'orange', 'green','magenta']

ax = ger_cars['company'].value_counts().plot(kind='bar' , figsize=(18,10),fontsize=20, color = color_list)
ax.set_xlabel("Car Company",fontsize=20,color='blue')
ax.set_ylabel("Total Number",fontsize=20,color='red')
plt.xticks(rotation="horizontal")
plt.suptitle('Count of German Car Companies', fontsize=30)
plt.show()
```
